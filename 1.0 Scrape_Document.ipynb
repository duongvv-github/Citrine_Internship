```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "http://iopscience.iop.org/article/10.1088/1367-2630/18/1/013003/meta"
resp = requests.get(url)
html = resp.text
```

```python
soup = BeautifulSoup(html, "lxml")
```

This finds all the tables and puts them into one variable, but I want each table individually

```python
table_1 = soup.find_all('table')[0]
df_1 = pd.read_html(str(table_1))
df_1
```

```python
table_2 = soup.find_all('table')[1]
df_2 = pd.read_html(str(table_2))
df_2
```
